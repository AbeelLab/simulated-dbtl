```python
def add_homoschedastic_noise(values,percentage):
    """add noise to the training set, to see the effect of noise models on performance"""
    mean_percentage=percentage
    newvalue=np.random.normal(values,mean_percentage)
    return newvalue
    
def add_heteroschedastic_noise(values,percentage):
    """add noise to the training set, to see the effect of noise models on performance"""
    mean_percentage=percentage*values
    value=np.random.normal(values,mean_percentage)
    return value
    

x=np.ones(100000)
y=np.ones(100000)*2
z=np.ones(100000)*3
a=np.ones(100000)*4
b=np.ones(100000)*5
c=np.ones(100000)*12
print(x)

x=add_heteroschedastic_noise(x,percentage=0.15)
y=add_heteroschedastic_noise(y,percentage=0.15)
z=add_heteroschedastic_noise(z,percentage=0.15)
a=add_heteroschedastic_noise(a,percentage=0.15)
b=add_heteroschedastic_noise(b,percentage=0.15)
c=add_heteroschedastic_noise(c,percentage=0.15)
sns.kdeplot(x)
sns.kdeplot(y)
sns.kdeplot(z)
sns.kdeplot(a)
sns.kdeplot(b)
sns.kdeplot(c)
plt.xlabel("Flux value")
plt.title("15% heteroschedastic noise")
```

```python

def pvalueheatmap(pairwise_list,algorithm,metric):       
    p_value=np.zeros((len(pairwise_list),len(pairwise_list)))
    for i in range(len(pairwise_list)):
        for j in range(len(pairwise_list)):
            x=ttest_ind(pairwise_list[i],pairwise_list[j])
            p_value[i,j]=x.pvalue
    fig1, axs = plt.subplots()
    fig1.tight_layout()
    plt.imshow(p_value,cmap="Reds")
    label_name=["no_noise","homo_04","homo_15","hetero_04","hetero_15"]
    axs.set_xticks(np.arange(np.shape(p_value)[0]))
    axs.set_yticks(np.arange(np.shape(p_value)[0]))
    axs.set_xticklabels(label_name)
    axs.set_yticklabels(label_name)

    plt.title("t-test "+ metric+" (p-value) "+algorithm+" (N=50)")
    for (j,i),label in np.ndenumerate(p_value):
        axs.text(i,j,np.round(label,3),ha='center',va='center')
        #axs.text(i,j,label,ha='center',va='center')

    return p_value


p_valueSGD=pvalueheatmap(SGD_pairwise_list,"Stochastic Gradient Descent","Intersect") 
#p_valueSGD=pvalueheatmap(SGD_pairwise_list2,"Stochastic Gradient Descent","R2") 


p_valueRF=pvalueheatmap(RF_pairwise_list,"Random Forest","Intersect") 
#p_valueRF=pvalueheatmap(RF_pairwise_list2,"Random Forest","R2") 


p_valueGBR=pvalueheatmap(GBR_pairwise_list,"Gradient Boosting","Intersect") 
#p_valueGBR=pvalueheatmap(GBR_pairwise_list2,"Gradient Boosting","R2") 

p_valueNN=pvalueheatmap(NN_pairwise_list,"Neural Network","Intersect") 
#p_valueNN=pvalueheatmap(NN_pairwise_list2,"Neural Network","R2") 
        
```
