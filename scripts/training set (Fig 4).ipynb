```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
dir1="..//results/no_noise_scenario_analysis/Figure3 training set size"

results50_r2=pd.read_csv(dir1+"/sc1_nonoise_50_3110.csv").iloc[:,1:]
results50_r2=results50_r2.drop(labels='GPR',axis=1)
results100_r2=pd.read_csv(dir1+"/sc1_nonoise_100_3110.csv").iloc[:,1:]
results100_r2=results100_r2.drop(labels='GPR',axis=1)
results200_r2=pd.read_csv(dir1+"/sc1_nonoise_200_3110.csv").iloc[:,1:]
results200_r2=results200_r2.drop(labels='GPR',axis=1)
results500_r2=pd.read_csv(dir1+"/sc1_nonoise_500_3110.csv").iloc[:,1:]
results500_r2=results500_r2.drop(labels='GPR',axis=1)
results1000_r2=pd.read_csv(dir1+"/sc1_nonoise_1000_3110.csv").iloc[:,1:]
results1000_r2=results1000_r2.drop(labels='GPR',axis=1)
results2000_r2=pd.read_csv(dir1+"/sc1_no_noise_r2_2000.csv").iloc[:,1:]
results5000_r2=pd.read_csv(dir1+"/sc1_no_noise_r2_5000.csv").iloc[:,1:]



results50_int=pd.read_csv(dir1+"/sc_equal_no_noise_intersect_50.csv").iloc[:,1:]
results100_int=pd.read_csv(dir1+"/sc_equal_no_noise_intersect_100.csv").iloc[:,1:]
results200_int=pd.read_csv(dir1+"/sc_equal_no_noise_intersect_200.csv").iloc[:,1:]
results500_int=pd.read_csv(dir1+"/sc_equal_no_noise_intersect_500.csv").iloc[:,1:]
results1000_int=pd.read_csv(dir1+"/sc_equal_no_noise_intersect_1000.csv").iloc[:,1:]
results2000_int=pd.read_csv(dir1+"/sc_equal_no_noise_intersect_2000.csv").iloc[:,1:]
results5000_int=pd.read_csv(dir1+"/sc_equal_no_noise_intersect_5000.csv").iloc[:,1:]


#order results in same way as R2 scores for plotting
results50_int=results50_int.iloc[:,[3,1,2,5,0,4,6]]
results100_int=results100_int.iloc[:,[3,1,2,5,0,4,6]]
results200_int=results200_int.iloc[:,[3,1,2,5,0,4,6]]
results500_int=results500_int.iloc[:,[3,1,2,5,0,4,6]]
results1000_int=results1000_int.iloc[:,[3,1,2,5,0,4,6]]
results2000_int=results2000_int.iloc[:,[3,1,2,5,0,4,6]]
results5000_int=results5000_int.iloc[:,[3,1,2,5,0,4,6]]
```

# Intersection benchmark
When doing multiple runs of the intersection, the intersection is not normally distributed. Therefore, we average each 10 runs and then take an average of that.

```python


x=list(results50_int.columns)
X_axis = np.arange(len(x))
y1=np.mean(results50_int)
y2=np.mean(results100_int)
y3=np.mean(results200_int)
y4=np.mean(results500_int)
y5=np.mean(results1000_int)

fig, ax1 = plt.subplots(figsize=(12, 4))
ax1.bar(X_axis-0.3,y1,width=0.15,align='center',label="50", yerr=np.std(results50_int),color="#00468BFF")
ax1.bar(X_axis-0.15,y2,width=0.15,align='center',label="100", yerr=np.std(results100_int),color="#ED0000FF")
ax1.bar(X_axis,y3,width=0.15,align='center',label="200", yerr=np.std(results200_int),color="#ADB6B6FF")
ax1.bar(X_axis+0.15,y4,width=0.15,align='center',label="500", yerr=np.std(results500_int),color="#925E9FFF")
ax1.bar(X_axis+0.3,y5,width=0.15,align='center',label="1000", yerr=np.std(results1000_int),color="#0099B4FF")
#ax1.bar(X_axis,0.4,y3)
plt.xticks(X_axis, x)
plt.legend(title="Training set size",bbox_to_anchor=(1, 1))
ax1.set_ylabel("Average top 100 prediction ")
ax1.set_xlabel("Algorithms")
ax1.set_title("Performance equal sampling scenario")

colors = ['#00468BFF', '#ED0000FF', '#42B540FF',"#0099B4FF",
         "#925E9FFF","#FDAF91FF","#AD002AFF","#ADB6B6FF"]
plt.savefig("../../../Results/Benchmark paper/Figure 3/sc1_trainingsetsize_intersection.svg",bbox_inches="tight")
```
